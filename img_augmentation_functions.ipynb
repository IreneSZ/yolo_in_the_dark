```python
from __future__ import print_function, division, absolute_import

import math
import numbers
import sys
import os
import cv2
import numpy as np

import matplotlib.pyplot as plt
```

```python
# flip before padding
# output image has the same shape as the input image
# bounding boxes also flipped
def horizontal_flip(img_path):
    img = cv2.imread(img_path)
    img_output = np.flip(img, 1)
    return img_output               
```

```python
# change the perspective of the object, vertically
```

```python
img_path = './image/n02777734_411.jpg'
img = cv2.imread(img_path)
```

```python
result = horizontal_flip(img_path)
```
